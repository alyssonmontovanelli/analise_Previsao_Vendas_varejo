#### IMPORTANDO O DF ALIMENTAÇÃO

```python
df_alimentacao = pd.read_csv("../data/dados_para_processamento/df_alimentacao.csv", sep =',', encoding = 'utf-8')
df_alimentacao = df_alimentacao[df_alimentacao['Data_Venda'] != '2020-01']
```

```python
df_alimentacao.sample(2)
```

```python
df_alimentacao['Produto'].value_counts()
```

```python
# Transformando valores de produtos em Dummies

df_alimentacao = pd.get_dummies(df_alimentacao, columns=['Produto'], drop_first=False)
```

```python
df_alimentacao.sample(5)
```

```python
variaveis_processadas = ['Produto_Cabo de Carregamento USB-C', 'Produto_Cabo de Carregamento Turbo', 'Produto_Pilhas AA (4 Unidades)', 'Produto_Pilhas AAA (4 Unidades)']

for x in variaveis_processadas:
    df_alimentacao[x] = np.where(df_alimentacao[x] == True, 1,0)
```

```python
df_alimentacao.sort_values(by = 'Data_Venda',ascending= True )
```
